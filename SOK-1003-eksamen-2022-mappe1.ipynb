## Oppgave 4 (20 poeng)

Et lån består som regel av et månedlig terminbeløp. Dette beløpet er summen av avdrag (nedbetalingen på lånet) og renter. Vi antar månedlig forrenting i alle oppgavene. Dvs. at det er 12 terminer i hvert år.<br><br>
a) Lag en funksjon som regner ut hvor mye lånet "x" koster deg i renteutgifter for "t" terminer med årlig rente "r" for et serielån. <br> <br>  Siden dette er et serielån, så vil avdragene være like hver måned men renteutgiftene reduseres i takt med avdragene. Renteutgiftene for en gitt termin "t" vil derfor være den årlige renten "r" (delt på antall forrentinger "f") på gjenværende beløp på det tidspunktet.   $renteutgifter_{t} = (x-a*(t-1))*{r/f}$   <br>  <br> 
Det vil si at renteutgiftene første termin er <br>
$renteutgifter_{1} = (x-a*0)*{r/f}$, og andre termin er <br>
$renteutgifter_{2} = (x-a*1)*{r/f}$ osv..<br> <br>
Siden vi er ute etter den totale kostnaden i svaret, må du summere renteutgiftene over alle terminer, det vil si $\sum_{t=1}^N(x-a*(t-1))*{r/f}$. Dette betyr egentlig bare $renteutgifter_{1} + renteutgifter_{2},...,+renteutgifter_{t}$


<br>
<b>Hint:</b> siden terminbeløpet varierer for hver måned (pga at rentene endres), må alle enkeltperioder summeres. Det kan være nyttige å bruke funksjonen np.arange() til dette. Mao, det er ikke nødvendig å bruke sigma ($\sum_{t=1}^N$) i formelen til dette 


```python
# added this just to have a cleaner print out
import pandas as pd
```

```python
def serial_loan(loan:int,rente:float,terminer_per_year:int,years:int,rente_format:str="desimal"):
    
    # check if input of percent is in desimal or percentage (0.04 or 4%)
    if rente_format == "percent" or rente_format == "%":
        rente = rente/100
    
    # init var
    terminer = terminer_per_year*years
    avdragsbeløp = loan/terminer
    rentesats  = rente/terminer_per_year
    rentebeløp = loan*rentesats
    rente_list = []
    termin_list = []
    rest_list = []
    
    # calculating
    for termin in range(1,terminer+1):
        current_value = ((loan - avdragsbeløp*(termin-1)))
        rente_list.append(current_value*rentesats)
        termin_list.append(rente_list[-1]+avdragsbeløp)
        rest_list.append((current_value*(1+rentesats))-(termin_list[-1]))
    
    # saving data
    data = {"Input Data": {"Loan": loan,
                           "Rente":f"{int((rente)*100)}%",
                           "Years":years,
                           "Term per year":terminer_per_year,
                           "Terminer": terminer,
                           "Rentesats":rentesats},
            "Cal":{"Avdrag":avdragsbeløp,
                   "Rente":rente_list,
                   "Termin":termin_list,
                   "Rest":rest_list},
            "Total":sum(termin_list)}
    return data

# call function
sl_data = serial_loan(2000000,4,12,20,"%")

#print info
for key,item in sl_data["Input Data"].items():
    print(f"{key} : {item}")
print(f"Total: {sl_data['Total']}\n")

#print function with panda
print(pd.DataFrame.from_dict(sl_data["Cal"]).round(5).astype(str))
```

c) Vis hva det samme lånet koster som annuitetslån, dvs differansen mellom alle terminbeløp og lånebeløp.<br> <br> Annuitetslån gir like terminbeløp hver måned, men renten utgjør en større del av dette beløpet i starten. Terminbeløpet for et annuitetslån er definert ved formelen:
$T = x*\frac{r/f} {(1-(1+(r/f))^{-t})}$, hvor x=lånebeløp, r = årlig rente, t = terminer, f= antall forrentinger





```python
def anu_loan(loan:int,rente:float,terminer_per_year:int,years:int,rente_format:str="desimal"):
    # check if input of percent is in desimal or percentage (0.04 or 4%)
    if rente_format == "percent" or rente_format == "%":
        rente = rente/100
    
    # initiation var
    terminer = terminer_per_year*years
    rentesats = rente/terminer_per_year
    constant_termin = loan*((rentesats)/(1-(1+(rentesats))**-terminer))
    rente_list = []
    termin_avdrag_list = []
    rest_list = [loan]
    current_rent_value = loan
    
    # calculating
    for termin in range(1,terminer+1):
        rente_list.append(rest_list[-1]*rentesats)
        termin_avdrag_list.append(constant_termin-rente_list[-1])
        rest_list.append(rest_list[-1]-termin_avdrag_list[-1])
    rest_list.pop(0)
    
    data = {"Input Data": {"Loan": loan,
                           "Rente":f"{int((rente)*100)}%",
                           "Years":years,
                           "Term per year":terminer_per_year,
                           "Terminer": terminer,
                           "Rentesats":rentesats},
            "Cal":{"Avdrag":termin_avdrag_list,
                   "Rente":rente_list,
                   "Termin":constant_termin,
                   "Rest":rest_list},
            "Total":sum(termin_avdrag_list)}

    return data

#print function with panda
for year in [10,20,30]:
    # print info and run function 
    print("-"*20,"\n")
    print(f"YEAR - {year}")
    anu_data = anu_loan(2000000,4,12,20,"%")
    
    # print input data
    print("\n-Input Data-\n")
    for key,item in anu_data["Input Data"].items():
        print(f"{key} : {item}")
    print(f"Total: {anu_data['Total']}\n")
    print("-"*20,"\n")
    
    # print calculations 
    print(pd.DataFrame.from_dict(anu_data["Cal"]).round(5).astype(str))
```
